```python
import os
from PIL import Image

def highlight_objects_in_directory(input_dir, output_dir):
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # List all files in the input directory
    image_files = [file for file in os.listdir(input_dir) if file.endswith('.png') or file.endswith('.jpg')]

    # Process each image in the input directory
    for filename in image_files:
        print("***************")
        print(filename)
        # Open the image
        image_path = os.path.join(input_dir, filename)
        image = Image.open(image_path)
        
        # Convert the image to grayscale if it's not already in 'L' mode
        if image.mode != 'L':
            image = image.convert('L')
        
        # Get the image size
        width, height = image.size
        #print(width)
        #print(height)
        # Define the threshold for brightness adjustment (you can adjust this value)
        threshold = 0

        label_set = set()
        
        # Process each pixel in the image
        for y in range(height):
            for x in range(width):
                # Get the pixel value (brightness)
                pixel = image.getpixel((x, y))
                
                # Check if the pixel is below the threshold (considered as object)
                #if pixel > threshold:
                # Increase the brightness of the pixel
                if pixel == 255 or pixel == 128:
                    new_pixel = 0
                elif pixel == 46:
                    new_pixel = 1
                elif pixel == 115:
                    new_pixel = 2
                elif pixel == 70:
                    new_pixel = 3
                elif pixel == 47:
                    new_pixel = 4
                elif pixel == 16:
                    new_pixel = 5
                elif pixel == 10:
                    new_pixel = 6
                elif pixel == 77:
                    new_pixel = 7
                elif pixel == 164:
                    new_pixel = 8
                elif pixel == 33:
                    new_pixel = 9
                elif pixel == 84:
                    new_pixel = 10
                else:
                    print("Anomaly: " + pixel)
                
                # if x == 96 and y == 188:
                #     print(pixel)
                    
                label_set.add(pixel)
                #new_pixel = 255  # You can adjust the value to control the brightness increase
                image.putpixel((x, y), new_pixel)
        
        # Save the modified image to the output directory
        print(label_set)
        pixels = list(image.getdata())

        # Find unique pixel values using a set
        unique_pixels = set(pixels)

        # Print the unique pixel values
        print("Unique new pixel values:", unique_pixels)
        
        
        output_path = os.path.join(output_dir, filename)
        image.save(output_path)

# Example usage
input_directory = 'img_1006/label_make_raw'
output_directory = 'img_1006/label_make'
highlight_objects_in_directory(input_directory, output_directory)
```
